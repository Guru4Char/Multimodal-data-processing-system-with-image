```python
# ==============================================================================
# 1. Setup and Installation
# ==============================================================================

# Install Tesseract OCR (the underlying engine)
print("Installing Tesseract OCR...")
!sudo apt update > /dev/null
!sudo apt install tesseract-ocr -y > /dev/null

# Install Python packages (pytesseract for the wrapper, pillow for image handling)
print("Installing Python dependencies...")
!pip install pytesseract pillow -q

# ==============================================================================
# 2. Imports and File Upload
# ==============================================================================

import pytesseract
from PIL import Image
from google.colab import files
import os

# Function to upload and process the file
def extract_text_from_uploaded_image():
    """Uploads an image file and performs OCR to extract text."""
    print("\nPlease upload an image file (e.g., PNG, JPG) to extract text from.")
    uploaded = files.upload()

    if not uploaded:
        print("No file uploaded. Exiting.")
        return

    # Get the file name of the uploaded file
    file_name = next(iter(uploaded))

    # Save the uploaded file locally (necessary for PIL to open it)
    with open(file_name, 'wb') as f:
        f.write(uploaded[file_name])

    print(f"\nProcessing '{file_name}'...")

    try:
        # Open the image using PIL (Pillow)
        img = Image.open(file_name)

        # ======================================================================
        # 3. Text Extraction (The Core OCR Call)
        # ======================================================================

        # Use pytesseract.image_to_string() to extract all text
        # The extracted text is returned as a single string
        extracted_text = pytesseract.image_to_string(img)

        # ======================================================================
        # 4. Output
        # ======================================================================

        print("\n" + "="*50)
        print("✅ EXTRACTED TEXT:")
        print("="*50)

        if extracted_text.strip():
            print(extracted_text)
        else:
            print("❌ No text could be extracted. Please check the image quality.")

    except Exception as e:
        print(f"❌ An error occurred during processing: {e}")

# Run the function
if __name__ == "__main__":
    extract_text_from_uploaded_image()
```
